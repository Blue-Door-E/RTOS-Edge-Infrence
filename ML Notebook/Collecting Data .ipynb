# Collecting data

This is just a raw notebook that can be ran to collect data on another machine. 

```python
import socket, struct, cv2, numpy as np, time, os
#.           "xxx.xxx.x.x"
HOST, PORT = "xxx.xxx.x.x", xxxx
SAVE_DIR = "dataset/raw/Walking around Work spin 2/"
os.makedirs(SAVE_DIR, exist_ok=True)

s = socket.socket()
s.connect((HOST, PORT))
print("Connected to stream")

count = 0
try:
    while True:
        hdr = s.recv(4)
        if len(hdr) < 4: break
        frame_len = struct.unpack(">I", hdr)[0]
        data = b""
        while len(data) < frame_len:
            pkt = s.recv(frame_len - len(data))
            if not pkt: break
            data += pkt
        img = cv2.imdecode(np.frombuffer(data, np.uint8), cv2.IMREAD_COLOR)
        filename = f"{SAVE_DIR}{int(time.time())}_{count}.jpg"
        cv2.imwrite(filename, img)
        count += 1
        if count % 50 == 0:
            print(f"Saved {count} frames...")
except KeyboardInterrupt:
    s.close()
    print("Stopped.")
```
